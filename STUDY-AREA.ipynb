```python
import geopandas as gpd
import matplotlib.pyplot as plt


# 1. LOAD SHAPEFILE
shapefile = r"C:/Users/ampon/Desktop/PYTHON/GHA_SHAPEFILE/gh_wgs16dregions.shp"

ghana = gpd.read_file(shapefile)

# 2. CREATE FIGURE
fig, ax = plt.subplots(figsize=(6,7))

# Plot regions
ghana.plot(
    ax=ax,
    edgecolor='black',
    cmap='Pastel1'
)

# 3. ADD REGION LABELS
for idx, row in ghana.iterrows():
    
    centroid = row.geometry.centroid
    
    ax.text(
        centroid.x,
        centroid.y,
        row['REGION'],
        fontsize=5.5,
        ha='center',
        fontweight='bold'
    )

ax.annotate(
    'N',
    xy=(0.95, 0.98),
    xytext=(0.95, 0.90),
    arrowprops=dict(facecolor='black', width=3, headwidth=10),
    ha='center',
    fontsize=14,
    xycoords=ax.transAxes
)


# 5. ADD TITLE AND AXIS LABELS
ax.set_title("STUDY AREA ", fontsize=15)

ax.set_xlabel("Longitude")
ax.set_ylabel("Latitude")

# Show grid for coordinates
ax.grid(True, linestyle="--", alpha=0.5)

# 6. DISPLAY MAP
plt.show()

```
