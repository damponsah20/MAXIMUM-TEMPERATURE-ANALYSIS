IMPORTING LIBRARIES

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from warnings import filterwarnings
filterwarnings('ignore')
```

DATA READING

```python
# Specify the path to your text file
file_path1 = 'Yendi_1960_2020_dTx_complete.txt'
file_path2 = 'Navrongo_1960_2020_dTx_complete.txt'
file_path3 = 'Wa_1960_2020_dTx_complete.txt'
file_path4 = 'Bole_1960_2020_dTx_complete.txt'
file_path5 = 'Walewale_1960_2020_dTx_complete.txt'
file_path6 = 'Tamale_1960_2020_dTx_complete.txt'
```

FIRST WINDOW

```python
# Read the file into a pandas DataFrame
yendi_1961_1990 = pd.read_csv(file_path1, delim_whitespace=True, header=None, names=["year", "month", "day", "Tmax"]).loc[366:11322] 
Navrongo_1961_1990 = pd.read_csv(file_path2, delim_whitespace=True, header=None, names=["year", "month", "day", "Tmax"]).loc[366:11322]  
Wa_1961_1990 = pd.read_csv(file_path3, delim_whitespace=True, header=None, names=["year", "month", "day", "Tmax"]).loc[366:11322] 
Bole_1961_1990 = pd.read_csv(file_path4, delim_whitespace=True, header=None, names=["year", "month", "day", "Tmax"]).loc[366:11322]  
Walewale_1961_1990 = pd.read_csv(file_path5, delim_whitespace=True, header=None, names=["year", "month", "day", "Tmax"]).loc[366:11322] 
Tamale_1961_1990 = pd.read_csv(file_path6, delim_whitespace=True, header=None, names=["year", "month", "day", "Tmax"]).loc[366:11322]
```

```python
Tamale_1961_1990
```

GROUPBY YEAR

```python
annual_yendi_1961_1990 = yendi_1961_1990.groupby('year')['Tmax'].mean().reset_index()
annual_Navrongo_1961_1990 = Navrongo_1961_1990.groupby('year')['Tmax'].mean().reset_index()
annual_Wa_1961_1990 = Wa_1961_1990.groupby('year')['Tmax'].mean().reset_index()
annual_Bole_1961_1990 = Bole_1961_1990.groupby('year')['Tmax'].mean().reset_index()
annual_Walewale_1961_1990 = Walewale_1961_1990.groupby('year')['Tmax'].mean().reset_index()
annual_Tamale_1961_1990 = Tamale_1961_1990.groupby('year')['Tmax'].mean().reset_index()
```

```python
monthly_yendi_1961_1990 = yendi_1961_1990.groupby('month')['Tmax'].mean().reset_index()
monthly_Navrongo_1961_1990 = Navrongo_1961_1990.groupby('month')['Tmax'].mean().reset_index()
monthly_Wa_1961_1990 = Wa_1961_1990.groupby('month')['Tmax'].mean().reset_index()
monthly_Bole_1961_1990 = Bole_1961_1990.groupby('month')['Tmax'].mean().reset_index()
monthly_Walewale_1961_1990 = Walewale_1961_1990.groupby('month')['Tmax'].mean().reset_index()
monthly_Tamale_1961_1990 = Tamale_1961_1990.groupby('month')['Tmax'].mean().reset_index()
```

ANOMALY CALCULATION


```python
# Calculate the overall mean and standard deviation for each location
overall_mean_tmax_yendi_1961_1990 = yendi_1961_1990['Tmax'].mean()
overall_std_tmax_yendi_1961_1990 = yendi_1961_1990['Tmax'].std()

overall_mean_tmax_Navrongo_1961_1990 = Navrongo_1961_1990['Tmax'].mean()
overall_std_tmax_Navrongo_1961_1990 = Navrongo_1961_1990['Tmax'].std()

overall_mean_tmax_Wa_1961_1990 = Wa_1961_1990['Tmax'].mean()
overall_std_tmax_Wa_1961_1990 = Wa_1961_1990['Tmax'].std()

overall_mean_tmax_Bole_1961_1990 = Bole_1961_1990['Tmax'].mean()
overall_std_tmax_Bole_1961_1990 = Bole_1961_1990['Tmax'].std()

overall_mean_tmax_Walewale_1961_1990 = Walewale_1961_1990['Tmax'].mean()
overall_std_tmax_Walewale_1961_1990 = Walewale_1961_1990['Tmax'].std()

overall_mean_tmax_Tamale_1961_1990 = Tamale_1961_1990['Tmax'].mean()
overall_std_tmax_Tamale_1961_1990 = Tamale_1961_1990['Tmax'].std()

# Calculate the anomaly for each year for each location
annual_yendi_1961_1990['Anomaly'] = (annual_yendi_1961_1990['Tmax'] - overall_mean_tmax_yendi_1961_1990) / overall_std_tmax_yendi_1961_1990
annual_Navrongo_1961_1990['Anomaly'] = (annual_Navrongo_1961_1990['Tmax'] - overall_mean_tmax_Navrongo_1961_1990) / overall_std_tmax_Navrongo_1961_1990
annual_Wa_1961_1990['Anomaly'] = (annual_Wa_1961_1990['Tmax'] - overall_mean_tmax_Wa_1961_1990) / overall_std_tmax_Wa_1961_1990
annual_Bole_1961_1990['Anomaly'] = (annual_Bole_1961_1990['Tmax'] - overall_mean_tmax_Bole_1961_1990) / overall_std_tmax_Bole_1961_1990
annual_Walewale_1961_1990['Anomaly'] = (annual_Walewale_1961_1990['Tmax'] - overall_mean_tmax_Walewale_1961_1990) / overall_std_tmax_Walewale_1961_1990
annual_Tamale_1961_1990['Anomaly'] = (annual_Tamale_1961_1990['Tmax'] - overall_mean_tmax_Tamale_1961_1990) / overall_std_tmax_Tamale_1961_1990

# Print the result for each location
print("Yendi Anomaly:\n", annual_yendi_1961_1990)
print("Navrongo Anomaly:\n", annual_Navrongo_1961_1990)
print("Wa Anomaly:\n", annual_Wa_1961_1990)
print("Bole Anomaly:\n", annual_Bole_1961_1990)
print("Walewale Anomaly:\n", annual_Walewale_1961_1990)
print("Tamale Anomaly:\n", annual_Tamale_1961_1990)
```
